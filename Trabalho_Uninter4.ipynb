```python
print("Bem vindo a Empresa de Logistica Tiago Ximenes Costa S.A.")
### FUNCAO DIMENSOESOBJETO ###
def dimensoesObjeto ():     #Funçao do tamanho do objeto
  while True:               # While para Loop a fim de o cliente colocar valores corretos
    try:
      altura = float(input("Insira o valor da altura do objeto: "))     #variavel para input de altura
      comprimento = float(input("Insira o valor do comprimento do objeto: "))  #variavel para input de comprimento
      largura = float(input("Insira o valor da largura do objeto: "))  #variavel para input de largura
      volume = (altura*comprimento*largura)  #formula de volume
      if volume >1 and volume <=1000 :   # Dependendo do volume, a funcao tera um valor especifico
        return 10
      elif volume>=1001 and volume<=10000 :
        return 20
      elif volume>=10001 and volume<=30000 :
        return 30
      elif volume>=30001 and volume<=100000 :
        return 50
      else:
        print("Volume não aceito. Tente Novamente")  #Caso o cliente coloque um numero diferente do esperado
        continue
    except ValueError:    #Caso o cliente coloque uma expressao que daria erro de valor
      print("Dimensão errada. Tente Novamente")
      continue
### FUNCAO PESOOBJETO ###
def pesoObjeto() :    #Funcao do peso do objeto
  while True:  #While para o cliente colocar o peso correto via loop
    try:
      peso= float(input("Insira o peso do objeto: ")) #Inserir o peso do objeto
      if peso > 0 and peso <= 0.1 :  #Conforme o peso, a funcao sai com um resultado diferente
        return 1
      elif peso <= 0.11 and peso <= 1 :
        return 1.5
      elif peso <= 1.10 and peso <= 10 :
        return 2
      elif peso <= 10.1 and peso <= 30:
        return 3
      else:
        print("Peso não aceito. Tente novamente")  #Caso o cliente coloque peso diferente do esperado
        continue
    except ValueError:  #Caso o cliente coloque um valor que daria erro no sistema
      print("Peso errado. Tente Novamente")
      continue

### FUNCAO ROTAOBJETO ###
def rotaObjeto() :   #Funcao para a rota do objeto 
  while True:   #loop para o cliente colocar a rota correta 
    try:
      print('''
      Insira a rota desejada:
      RS -> Rio de Janeiro até São Paulo
      SR -> São Paulo até Rio de Janeiro
      BS -> Brasília até São Paulo
      SB -> São Paulo até Brasília
      BR -> Brasília até Rio de Janeiro
      RB -> Rio de Janeiro até Brasília
      ''')  #Texto com rotas disponiveis
      rota = input("Insira a rota : ")  #Para o cliente colocar a rota correta
      if rota == "RS" :  #Opcoes que, dependendo da rota, fazem sair um resultado da funcao diferente
        return 1
      elif rota == "SR" :
        return 1
      elif rota == "BS" :
        return 1.2
      elif rota == "SB" :
        return 1.2
      elif rota == "BR" :
        return 1.5
      elif rota == "RB" :
        return 1.5
      else :   #Caso o cliente coloque valor fora do esperado
        print("Rota inexistente. Tente Novamente")
        continue
    except ValueError: #Caso o cliente coloque valor de Rota que daria erro no sistema
      print("Rota errada. Tente novamente")
      continue

### PRINCIPAL ###
dimensao = dimensoesObjeto()  #variavel que determina a funcao da dimensao
pesototal = pesoObjeto()  #variavel que determina a funcao de peso
rotaentrega = rotaObjeto() #variavel que determina a funcao da rota
cod3818596= dimensao * pesototal * rotaentrega  #variavel que determina o valor total do pedido
print("Obrigado pelo seu pedido")
print("O total a pagar é: R$ {:.2f}". format(cod3818596))  #mostragem do valor total


```

```python
listaPecas= []

################INÍCIO CADASTRAR PEÇA#########################

def cadastrarPeca( cp ) :
  print("Você está na etapa de cadastrar peça")
  print("O código da peça que está sendo cadastrada é: {}" .format(cp))
  nome = input("Insira o nome da peça: ")
  fabricante = input("Insira o fabricante da peça: ")
  valor = input("Insira o valor da peça: ")
  dicionarioPeca = {'código peça': cp, 
                    'nome'       : nome,
                    'fabricante' : fabricante,
                    'valor'      : valor }
  listaPecas.append(dicionarioPeca.copy())

################FIM CADASTRAR PEÇA############################
################INÍCIO CONSULTAR PEÇA#########################

def consultarPeca() :
  while True :
    try:
      print('''
  Você está na etapa de consultar peça

  As funções disponíveis são:

  Tecla:                Função:
  1............Consultar todas as peças
  2.......... Consultar peças por código
  3.........Consultar peças por fabricante
  4.....................Retornar
  ''')
      opConsultar = int(input("Insira a tecla referente a função desejada: "))
      if opConsultar == 1 :
        print("Você digitou a função para consultar todas as peças")
        for peca in listaPecas:
          for key, value in peca.items():
            print("{} : {}" .format(key, value))
      elif opConsultar == 2 :
        print("Você digitou a função para consultar peças por código")
        cod3818596= int(input("Digite o código da peça"))
        for peca in listaPecas:
          if(peca['cp'] == cod3818596):
            for key, value in peca.items():
              print("{} : {}" .format(key, value))
      elif opConsultar== 3 :
        print("Você digitou a função para consultar peças por fabricante")
        cod3818596= input("Digite o fabricante da peça")
        for peca in listaPecas:
          if(peca['fabricante'] == cod3818596):
            for key, value in peca.items():
              print("{} : {}" .format(key, value))
      elif opConsultar== 4 :
        print("Você digitou a função para retornar")
        break
      else:
        print("Esta tecla não tem função definida. Digite apenas as teclas mencionadas")
        continue
    except ValueError:
      print("Esta tecla não tem função definida e não é um número inteiro. Digite apenas as teclas mencionadas")
  
  
################FIM CONSULTAR PEÇA############################
################INÍCIO REMOVER PEÇA###########################

def removerPeca() :
  print("Você está na etapa de remover peça")
  cod3818596= int(input("Digite o código da peça"))
  for peca in listaPecas:
    if(peca['cp'] == cod3818596):
      listaPecas.remove(peca)

################FIM REMOVER PEÇA##############################
###################INÍCIO PRINCIPAL###########################

print("Bem vindo ao estoque virtual da Bicicletaria de Tiago Ximenes Costa")
print('''
Tecla:                   Função:
  1..................Cadastrar Peça
  2..................Consultar Peça
  3...................Remover Peça
  4.......................Sair
''')

codigoPeca= 0

while True:
  try:
    opcao=int(input("Digite a tecla da função desejada: "))
    if opcao == 1 :
      codigoPeca+= 1
      cadastrarPeca(codigoPeca)
    elif opcao ==2 :
      consultarPeca()
    elif opcao ==3 :
      removerPeca()
    elif opcao ==4 :
      print("Estoque virtual fechado. Saindo do programa.")
      break
    else :
      print("Esta tecla não tem função definida. Digite apenas as teclas mencionadas")
  except ValueError:
    print("Esta tecla não tem função definida e não é um número inteiro. Digite apenas as teclas mencionadas")

#################FIM PRINCIPAL#################################
```
